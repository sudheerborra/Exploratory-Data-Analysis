## 2.  Univariate Analysis

```python
#Let us take the variable age and perform the analysis

#plot the pdf of Age. This tells us what age people are more affected by cancer

'''we can visualize the percentage of people affected by cancer with age. We can find that number of people with diagnosed with
cancer decreses after the age 0f 65'''


counts, bin_edges=np.histogram(data_haberman['Age'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.figure
plt.title("PDF and CDF of Age")
plt.xlabel('Age')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);
plt.show()
```

##### There are more people between 50 and 55 age affected by cancer

#### Survival with Age

```python
import numpy as np

Age_Surv_1=data_haberman.loc[data_haberman["Survival_Status"] == 1]
Age_Surv_2=data_haberman.loc[data_haberman["Survival_Status"] == 2]

counts, bin_edges=np.histogram(Age_Surv_1['Age'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.figure(figsize=(8,4))
plt.subplot(121)
plt.title('PDF and CDF of Survial_Status_1')
plt.xlabel('Age')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);


counts, bin_edges=np.histogram(Age_Surv_2['Age'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.figure(figsize=(8,4))
plt.subplot(122)
plt.title('PDF and CDF of Survival_Status_2')
plt.xlabel('Age')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
```

###### From fig 2 above, it is observed that people affected by cancer at between the ages 40 and 50 are more likely to have Surv_Status 2

### Survival with axil_nodes

```python
Age_Surv_1=data_haberman.loc[data_haberman["Survival_Status"] == 1]
Age_Surv_2=data_haberman.loc[data_haberman["Survival_Status"] == 2]

counts, bin_edges=np.histogram(Age_Surv_1['Axil_Nodes'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.figure(figsize=(8,4))
plt.subplot(121)
plt.title('Survial_Status_1')
plt.xlabel('Axil_Nodes')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);


counts, bin_edges=np.histogram(Age_Surv_2['Axil_Nodes'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.figure(figsize=(8,4))
plt.subplot(122)
plt.title('Survival_Status_2')
plt.xlabel('Axil_Nodes')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf);
```

### More the number of Axil Nodes, less the chance of survival
### More than 80% of patients has lessthan 4 Axil_Nodes
